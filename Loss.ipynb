```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
Epochs = 5000
```

```python
path = './AE_in_32x32x2_out_32x32x1'
```

# Feature Dimension

```python
nf = [8, 16, 32]

for Horizon in [0, 60, 120, 180, 240]:
    for Smooth in [4, 16, 32]:
        try:
            plt.figure()
            plt.title('Horizon:{} Smooth:{}'.format(Horizon, Smooth))
            plt.xlabel('Iterations')
            plt.ylabel('Loss')

            for i in range(len(nf)):
                loss = np.load(path+'/Epochs_{0}_Horizon_{1}_Smooth_{2}_nf_{3}/Trainloss.npy'.format(Epochs, Horizon, Smooth, nf[i]))
                plt.plot(loss[:], label='nf={}'.format(nf[i]))
                plt.legend()

            plt.savefig('loss_32_32/Horizon:{} Smooth:{}.png'.format(Horizon, Smooth), dpi=300)
            plt.close()
        except:
            plt.close()
```

# Smooth

```python
Smooth = [4, 16, 32]

for Horizon in [0, 60, 120, 180, 240]:
    for nf in [8, 16, 32]:
        try:
            plt.figure()
            plt.title('Horizon:{} nf:{}'.format(Horizon, nf))
            plt.xlabel('Iterations')
            plt.ylabel('Loss')

            for i in range(len(Smooth)):
                loss = np.load(path+'/Epochs_{0}_Horizon_{1}_Smooth_{2}_nf_{3}/Trainloss.npy'.format(Epochs, Horizon, Smooth[i], nf))
                plt.plot(loss[:], label='Smooth={}'.format(Smooth[i]))
                plt.legend()

            plt.savefig('loss_32_32/Horizon:{} nf:{}.png'.format(Horizon, nf), dpi=300)
            plt.close()
        except:
            plt.close()
```

# Horizon

```python
Horizon = [0, 60, 120, 180, 240]

for Smooth in [4, 16, 32]:
    for nf in [8, 16, 32]:
        try:
            plt.figure()
            plt.title('nf:{} Smooth:{}'.format(nf, Smooth))
            plt.xlabel('Iterations')
            plt.ylabel('Loss')

            for i in range(len(Horizon)):
                loss = np.load(path+'/Epochs_{0}_Horizon_{1}_Smooth_{2}_nf_{3}/Trainloss.npy'.format(Epochs, Horizon[i], Smooth, nf))
                plt.plot(loss[:], label='Horizon={}'.format(Horizon[i]))
                plt.legend()

            plt.savefig('loss_32_32/Smooth:{} nf:{}.png'.format(Smooth, nf), dpi=300)
            plt.close()
        except:
            plt.close()
```

# Together

```python
plt.figure(dpi=300)
plt.title('Total Comparison')
plt.xlabel('Iterations')
plt.ylabel('Loss')

try:
    for Smooth in [4, 16, 32]:
        for nf in [8, 16, 32]:
            for Horizon in [0, 60, 120, 180, 240]:
                loss = np.load(path+'/Epochs_{0}_Horizon_{1}_Smooth_{2}_nf_{3}/Trainloss.npy'.format(Epochs, Horizon, Smooth, nf))
                plt.plot(loss[:], label='Horizon={} Smooth={} nf={}'.format(Horizon, Smooth, nf))
                plt.legend(fontsize=4)
except:
    pass

plt.savefig('loss_32_32/Total.png', dpi=300)
plt.show()
plt.close()
        
```
